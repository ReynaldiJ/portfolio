```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#generate data
from sklearn.datasets import make_blobs

x, y_true = make_blobs(n_features = 2, n_samples = 300, centers = 3, cluster_std = 1.5)

fig, ax = plt.subplots()
sns.scatterplot(ax=ax, x=x[:,0], y=x[:,1], hue=y_true)
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.title('Generated Data Scatter Plot')
plt.xlabel('X1 axis')
plt.ylabel('X2 axis')
plt.show()
```

```python
#unsupervised model
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(8)
fig.set_figheight(4)
sns.scatterplot(ax=ax1, x=x[:,0], y=x[:,1], hue=y_true)
sns.scatterplot(ax=ax2, x=x[:,0], y=x[:,1], hue=y_kmeans)
ax1.set_xlim(-15,15)
ax1.set_ylim(-15,15)
ax2.set_xlim(-15,15)
ax2.set_ylim(-15,15)
ax1.set_title('Generated Data Scatter Plot')
ax1.set_xlabel('X1 axis')
ax1.set_ylabel('X2 axis')
ax2.set_title('Generated Data Scatter Plot')
ax2.set_xlabel('X1 axis')
ax2.set_ylabel('X2 axis')
plt.show()
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
#calculate dist
from scipy.spatial.distance import cdist

# print(cdist(x[y_kmeans==0], [centers[0]]).max())

radius = [cdist(x[y_kmeans==i], centers).max() for i, center in enumerate(centers)]
radius
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(8)
fig.set_figheight(4)
sns.scatterplot(ax=ax1, x=x[:,0], y=x[:,1], hue=y_true)
sns.scatterplot(ax=ax2, x=x[:,0], y=x[:,1], hue=y_kmeans)
sns.scatterplot(ax=ax2, x=centers[:,0], y=centers[:,1], c='black', s=200, alpha=0.5)
for c,r in zip(centers,radius):
  ax2.add_patch(plt.Circle(c,r, fc='blue', alpha=0.1))
ax1.set_xlim(-15,15)
ax1.set_ylim(-15,15)
ax2.set_xlim(-15,15)
ax2.set_ylim(-15,15)
ax1.set_title('Generated Data Scatter Plot')
ax1.set_xlabel('X1 axis')
ax1.set_ylabel('X2 axis')
ax2.set_title('Generated Data Scatter Plot')
ax2.set_xlabel('X1 axis')
ax2.set_ylabel('X2 axis')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN as DBS
from sklearn.datasets import make_moons
```

```python
x, y_true = make_moons(200, noise = .05)

kmeans = KMeans(n_clusters=2)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
kmeans.inertia_

fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(8)
fig.set_figheight(4)
sns.scatterplot(ax=ax1, x=x[:,0], y=x[:,1], hue=y_true)
sns.scatterplot(ax=ax2, x=x[:,0], y=x[:,1], hue=y_kmeans)
ax1.set_title('Generated Data Scatter Plot')
ax1.set_xlabel('X1 axis')
ax1.set_ylabel('X2 axis')
ax2.set_title('Generated Data Scatter Plot KMeans')
ax2.set_xlabel('X1 axis')
ax2.set_ylabel('X2 axis')
plt.show()
```

```python
dbscan = DBS(eps=0.2).fit(x)
y_dbscan = dbscan.labels_

fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(8)
fig.set_figheight(4)
sns.scatterplot(ax=ax1, x=x[:,0], y=x[:,1], hue=y_kmeans)
sns.scatterplot(ax=ax2, x=x[:,0], y=x[:,1], hue=y_dbscan)
ax1.set_title('Generated Data Scatter Plot KMeans')
ax1.set_xlabel('X1 axis')
ax1.set_ylabel('X2 axis')
ax2.set_title('Generated Data Scatter Plot DBSCAN')
ax2.set_xlabel('X1 axis')
ax2.set_ylabel('X2 axis')
plt.show()
```

```python
x, y_true = make_blobs(n_features = 2, n_samples = 300, centers = 4, cluster_std = 1.2)
rdm = np.random.RandomState(13)
x_stretch = np.dot(x, rdm.randn(2,2))

kmeans = KMeans(n_clusters=4)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
kmeans.inertia_

fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(8)
fig.set_figheight(4)
sns.scatterplot(ax=ax1, x=x_stretch[:,0], y=x_stretch[:,1], hue=y_true)
sns.scatterplot(ax=ax2, x=x_stretch[:,0], y=x_stretch[:,1], hue=y_kmeans)
ax1.set_xlim(-15,15)
ax1.set_ylim(-15,15)
ax2.set_xlim(-15,15)
ax2.set_ylim(-15,15)
ax1.set_title('Generated Data Scatter Plot')
ax1.set_xlabel('X1 axis')
ax1.set_ylabel('X2 axis')
ax2.set_title('Generated Data Scatter Plot')
ax2.set_xlabel('X1 axis')
ax2.set_ylabel('X2 axis')
plt.show()
```

```python
gaussian = GaussianMixture(n_components=4)
gaussian.fit(x_stretch)
y_gaussian = gaussian.predict(x_stretch)

fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figwidth(8)
fig.set_figheight(4)
sns.scatterplot(ax=ax1, x=x_stretch[:,0], y=x_stretch[:,1], hue=y_kmeans)
sns.scatterplot(ax=ax2, x=x_stretch[:,0], y=x_stretch[:,1], hue=y_gaussian)
ax1.set_title('Generated Data Scatter Plot')
ax1.set_xlabel('X1 axis')
ax1.set_ylabel('X2 axis')
ax2.set_title('Generated Data Scatter Plot')
ax2.set_xlabel('X1 axis')
ax2.set_ylabel('X2 axis')
plt.show()
```
